Getting from repo

```python
headers = {'X-GitHub-Api-Version': '2022-11-28'}
```

```python
api_url_base = 'https://api.github.com'
owner = 'amzn' #username
url = f'{api_url_base}/users/{owner}/repos'
```

```python
url
```

```python
response = requests.get(url, headers=headers)
response.status_code
```

```python
len(response.json())
```

API Auth

```python

from AuthToken import TokenProvider

headers = {'Authorization': 'Bearer ' + TokenProvider.TOKEN, 
           'X-GitHub-Api-Version': '2022-11-28'}
```

Pagination

```python
repos_list = []
for page_num in range(1, 6): # get all the pages from amazon repos
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.extend(response.json())
    except:
        repos_list.append(None)
```

```python
len(repos_list) # successfuly getting the repos in all pages
```
